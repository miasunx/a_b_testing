Following is the code for t-test. Pay attention to the p-values

```python
from scipy.stats import ttest_ind
ind_t_test=ttest_ind(sample_control1, sample_treated1)
p_val = ind_t_test[1]
diff_mean = sample_control1.mean() - sample_treated1.mean()

output = "The results of the independent t-test are: \n\tdifference in mean = {%r}\n\tp-value = {%r} "%(float(diff_mean),float(p_val))
print(output)
alpha = 0.05


if p_val > alpha:
	print('Cannot reject the null hypothesis that the means are equal. Difference between treated and control is not statistically significant')
else:
	print('Reject the null hypothesis that the means are equal. Difference between treated and control is statistically significant')
```
